# Eigen Faces

This notebook was written to accompany the following article on EigenFaces.

### Download & Load the dataset 

```python
from sklearn.datasets import fetch_lfw_people

# Download the data, if not already on disk and load it as numpy arrays

lfw_people = fetch_lfw_people(min_faces_per_person=70)

# introspect the images arrays to find the shapes (for plotting)
n_samples, h, w = lfw_people.images.shape

print("number of samples: ",n_samples)
print("image dimensions: ", h, w)
print("number of pixels per image: ", h*w)
```

```python
print(lfw_people.target_names)
```

```python
# Import basic libraries & define helper functions
import numpy as np
import cv2
import matplotlib.pyplot as plt
from sklearn.cross_validation import train_test_split
from sklearn.decomposition import PCA
from sklearn.linear_model import SGDClassifier
from sklearn import metrics

%matplotlib inline  
```

```python
# Simple helper function to display a list of images with titles 
def show_images(images, cols = 1, titles = None):
    if not ((titles is None)or (len(images) == len(titles))):
        print("Ensure that the number of images matches number of titles")
        return
    n_images = len(images)
    if titles is None: titles = ['Image (%d)' % i for i in range(1,n_images + 1)]
    fig = plt.figure()
    for n, (image, title) in enumerate(zip(images, titles)):
        a = fig.add_subplot(cols, np.ceil(n_images/float(cols)), n + 1)
        if image.ndim == 2:
            plt.gray()
        plt.imshow(image.reshape((h, w)),cmap=plt.cm.gray)
        a.set_title(title,fontsize=20)
    fig.set_size_inches(np.array(fig.get_size_inches()) * n_images)
    plt.show()
```

### Prepare dataset for training

```python
# Prepare training and testing splits
X = lfw_people.data
Y = lfw_people.target
classes = lfw_people.target_names

X_train, X_test, y_train,y_test = train_test_split(X,Y,test_size=0.3)
```

```python
# Mean face (calculated among the training images)
mu = np.mean(X_train,0)

#display the mean face along with some sample faces from the dataset
show_images(images = [mu,X_train[0],X_train[1]],titles = ['mean','Sample 1','Sample 2'])
```

```python
# Subtract each face by the mean
X_train_m = np.zeros(X_train.shape)
X_train_m = X_train-mu

# check a sample pair
show_images([X_train[100],X_train_m[100]], titles=["original","after subtracting mean"])

# Apply the same transform to test data
X_test_m = np.zeros(X_test.shape)
X_test_m = X_test-mu
```

### EigenFaces

```python
# singular value decomposition (PCA)
U, S, V = np.linalg.svd(X_train_m.transpose(), full_matrices=False)
```

```python
# plot the top 10 eigenfaces
e_faces = U.transpose()[:10]
show_images(e_faces, 2)
```

```python
# weights is an n x n matrix
weights = np.dot(X_train_m, U)
print("original number of features: ", h*w)
print("reduced number of features: ",weights.shape[1])
```

```python
# reconstruct the face located at img_idx
img_idx = 50
recon = mu + np.dot(weights[img_idx, :], U.T)
show_images([X_train[img_idx],recon],titles = ["original", "reconstructed"])
```

```python
# Reconstructing using only a smaller subset of the weights learnt with PCA
img_idx = 50
reconstructed = []
labels = []
for k in range(50, weights.shape[1],100):
    recon = mu + np.dot(weights[img_idx, 0:k], U[:, 0:k].T)
    reconstructed.append(recon)
    labels.append("%d eigenfaces"%k)
show_images(images = reconstructed, cols=2, titles=labels)
```

### Classification & Evaluation

```python
n_components = 190
# We use sklearn's PCA classifier for simplicity
pca = PCA(n_components=n_components, whiten=True)
pca.fit(X_train_m)
```

```python
# Plot the importance of the eigenfaces based on their eigenvalues 
# (higher the eigen value, the more important it is)
f,ax = plt.subplots()
plt.plot(pca.explained_variance_ratio_)
plt.title("ratio of explained variance")
ax.set_xlabel("principal component")
plt.show()
```

```python
eigenfaces = pca.components_

#plot the top 10 eigenfaces (since most of the variance is captured by them)
show_images(eigenfaces[:10])
```

```python
# Reconstruct the image at img_idx using the PCA weights learned
img_idx = 50
x_pca = pca.transform([X_train_m[img_idx]])
x_recon = pca.inverse_transform([x_pca])
show_images([X_train[img_idx],x_recon],titles = ["original", "reconstructed"])
```

```python
# Apply the pca transformation on the training and testing data

X_train_pca = pca.transform(X_train_m)
X_test_pca = pca.transform(X_test_m)
```

```python
# Train a simple linear model on the reduced features

print("Fitting the classifier to the training set")
clf = SGDClassifier()
clf = clf.fit(X_train_pca, y_train)
```

```python
# Evaluate Training accuracy
y_train_pred = clf.predict(X_train_pca)
print ("Training accuracy: ",metrics.accuracy_score(y_train, y_train_pred))

# Evaluate Testing accuracy
y_test_pred = clf.predict(X_test_pca)
print ("Testing accuracy:",metrics.accuracy_score(y_test, y_test_pred))

# Get more detailed classifcation scores
print(" \n######### Classification report for Test ###### ")
print(metrics.classification_report(y_test,y_test_pred, target_names=classes))
```
